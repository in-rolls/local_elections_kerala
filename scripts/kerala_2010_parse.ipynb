```python
import os
import pandas as pd
from bs4 import BeautifulSoup
```

```python
folder_path = 'data_2010/downloaded_htmls/'

for filename in os.listdir(folder_path):
    if filename.endswith('.html'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            soup = BeautifulSoup(file, 'html.parser')
            
            # Find the specific div block containing the table
            div_block = soup.find('div', id='block-zircon-content')
            if div_block:
                # Find the first table within the div
                table = div_block.find('table')
                if table:
                    # Convert the table to a DataFrame
                    df = pd.read_html(str(table))[0]
                    dataframes.append(df)
                else:
                    print(f"No table found in the specified div in file: {filename}")
            else:
                print(f"No specified div block found in file: {filename}")

# Concatenate all DataFrames if there are any
if dataframes:
    final_df = pd.concat(dataframes, ignore_index=True)
    final_df.to_csv('combined_tables.csv', index=False)
    print("Tables have been successfully extracted and concatenated into 'combined_tables.csv'")
else:
    print("No tables were found in any of the HTML files.")
```
